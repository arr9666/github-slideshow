# Artifical Neural Network

```
# basic imports 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('diabetes.csv')
```

```
data.head()
```

```
data.describe()
```

```
X = data.iloc[:,:-1]
X
```

```
y = data.iloc[:,-1]
y
```

```
from keras import Sequential  # improt ann
```

```
from keras.layers import Dense
```

```
import keras
```

```
model = Sequential()  # initialize model
```

```
# adding layers
model.add(Dense(units=8, input_dim =8)) # first layer
model.add(Dense(units=12, activation='tanh')) # Second layer # activations = relu, tanh
model.add(Dense(units=1, activation='sigmoid')) # last layer # activations = softmax, sigmoid
```

```
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy']) # sgd stochastic gradient descident

#keras.optimizers.Adam(learning_rate=0.01)
```

```
# train

model.fit(X,y , epochs= 150, batch_size=10, verbose=1)
```
